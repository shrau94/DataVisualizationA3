```
import pandas as pd
```

```
taxi_zones = pd.read_csv('/content/drive/My Drive/DV_data/taxi+_zone_lookup.csv', usecols=['LocationID','Borough'])
green_taxis = pd.read_csv('/content/drive/My Drive/DV_data/green_tripdata_2019-12.csv')
yellow_taxis = pd.read_csv('/content/drive/My Drive/DV_data/yellow_tripdata_2019-12.csv')
```

```
taxi_zones_dict = taxi_zones.set_index('LocationID')['Borough'].to_dict()
green_taxis
```

```
green_taxi_replaced = green_taxis.replace({'PULocationID': taxi_zones_dict}).replace({'DOLocationID': taxi_zones_dict})
```

```
green_taxi_replaced
```

```
green_taxi_replaced.rename(columns={"PULocationID": "source", "DOLocationID": "target"}).groupby(["source", "target"]).size().to_frame(name='value').reset_index().to_json(orient='records')
```

```
green_taxi_replaced.rename(columns={"PULocationID": "source", "DOLocationID": "target"}).groupby(["source", "target"]).size()
```

```
def convert_timestamp_to_date(x):
  return x.split(" ")[0]
```

```
green_taxis['start_time'] = green_taxis['lpep_pickup_datetime'].apply(convert_timestamp_to_date)
```

```
green_taxis["start_time"].unique()
```

```
start_time_green = green_taxis.groupby("start_time")['trip_distance', 'passenger_count', 'fare_amount', 'tip_amount'].sum().reset_index()
```

```
def get_year(x):
  return x.split("-")[0]
  
start_time_green["year"] = start_time_green['start_time'].apply(get_year)
```

```
indexNames = start_time_green[(start_time_green['year'] != '2019') | (start_time_green['start_time'] == '2019-11-30')].index
 
start_time_green.drop(indexNames , inplace=True)
start_time_green.reset_index()
```

```
start_time_green.to_csv('/content/drive/My Drive/DV_data/green_taxi_analysis.csv')
```

```
start_time_green.to_json(orient='records')
```
